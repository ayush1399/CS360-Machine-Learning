### Part 1: Calculate the Mean, Median, Mode and Standard deviation

1. Load the wine dataset from the given 'wine.csv' file into a variable (say data).

```
import pandas as pd

data = pd.read_csv('wine.csv')
```

2. From the variable, extract the numerical information into a new variable, say scores.

```
import numpy as np

scores = data.to_numpy()
```

3. Calculate the row-wise, column-wise, and overall mean, median, mode, and standard deviation of the dataset.

```
row_wise_mean = np.mean(scores, axis=1)
column_wise_mean = np.mean(scores, axis=0)
overall_mean = np.mean(scores)

print(f"ROW_WISE_MEAN\n{row_wise_mean}")
print(f"COLUMN_WISE_MEAN\n{column_wise_mean}")
print(f"OVERALL_MEAN: {overall_mean}")

row_wise_median = np.median(scores, axis=1)
column_wise_median = np.median(scores, axis=0)
overall_median = np.median(scores)

print(f"ROW_WISE_MEDIAN\n{row_wise_median}")
print(f"COLUMN_WISE_MEDIAN\n{column_wise_median}")
print(f"OVERALL_MEDIAN: {overall_median}")

row_wise_sd = np.std(scores, axis=1)
column_wise_sd = np.std(scores, axis=0)
overall_sd = np.std(scores)
print(f"ROW_WISE_STANDARD_DEVIATION\n{row_wise_sd}")
print(f"COLUMN_WISE_STANDARD_DEVIATION\n{column_wise_sd}")
print(f"OVERALL_STANDARD_DEVIATION: {overall_sd}")

from scipy.stats import mode

row_wise_mode = mode(scores, axis=1).mode
column_wise_mode = mode(scores, axis=0).mode
overall_mode = mode(scores, axis=None).mode

print(f"ROW_WISE_MODE\n{row_wise_mode}")
print(f"COLUMN_WISE_MODE\n{column_wise_mode}")
print(f"OVERALL_MODE: {overall_mode}")
```

### Part 2: Calculate the pearson correlation and covariance matrix

1. Take the 'alcohol' and 'Proline' as the features from the 'wine.csv' dataset and calculate the pearson correlation for the same.

```
alcohol = np.reshape(data['Alcohol'].to_numpy(), newshape=(1, -1))
proline = np.reshape(data['Proline'].to_numpy(), newshape=(1, -1))

matrix1 = np.concatenate((alcohol, proline), axis=0)

pearson_correlation = np.corrcoef(matrix1)
print(f"PEARSON_CORRELATION\n{pearson_correlation}")
```

2. Take any 5 features from the 'wine.csv' datasets and load them in a matrix. Thereafter, create the covariance matrix of the same.


```
phenol = np.reshape(data['Phenols'].to_numpy(), newshape=(1, -1))
flavanoids = np.reshape(data['Flavanoids'].to_numpy(), newshape=(1, -1))
od = np.reshape(data['OD'].to_numpy(), newshape=(1, -1))

matrix2 = np.concatenate((alcohol, phenol, proline, flavanoids, od), axis=0)

covariance_matrix = np.cov(matrix2)
print(f"COVARIANCE MATRIX:\n {covariance_matrix}")
```
